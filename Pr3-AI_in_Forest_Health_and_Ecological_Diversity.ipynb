```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/forest_health_data_with_target.csv')

# Convert 'Health_Status' to categorical
df['Health_Status'] = df['Health_Status'].astype('category')

# Display dataset structure
print("Data Types:\n", df.dtypes)
print("\nFirst 5 Rows:\n", df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Countplot of health status
plt.figure(figsize=(7, 5))
sns.countplot(data=df, x='Health_Status', palette='Set2')
plt.title("Distribution of Tree Health Status")
plt.xlabel("Health Status")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Health_Status', y='Temperature', palette='coolwarm')
plt.title("Temperature vs Tree Health Status")
plt.show()
```

```python
soil_features = ['Soil_TN', 'Soil_TP', 'Soil_AP', 'Soil_AN']

for feature in soil_features:
    plt.figure(figsize=(8, 4))
    sns.boxplot(data=df, x='Health_Status', y=feature, palette='YlGn')
    plt.title(f'{feature} vs Health Status')
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Elevation', y='Tree_Height', hue='Health_Status', palette='Dark2')
plt.title("Tree Height vs Elevation (colored by Health Status)")
plt.xlabel("Elevation")
plt.ylabel("Tree Height")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, x='Health_Status', y='Fire_Risk_Index', palette='OrRd')
plt.title("Fire Risk Index by Tree Health Status")
plt.show()
```

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```
